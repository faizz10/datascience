```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
values = np.random.randn(100)
s = pd.Series(values)
print("series\n",s)
s.plot(kind='hist', title='Normally distributed random values')
plt.show()
```
